#**Leyendo paginas HTML y XML**

##importando paginas web


```python
import pandas as pd

datos_html = pd.read_html('peliculas_wikipedia.html')
datos_html
```

```python
#EL CONTENIDO DE LOS DATOS HTML ES UNA LISTA QUE CONTIENE 3 TABLAS

type(datos_html)
```

```python
len(datos_html)
```

```python
#SELECCIONAMOS LA TABLA DE LA SGUNDA POSICION QUE VAMOS A UTILIZAR

top_peliculas = datos_html[1]

top_peliculas
```

##Escribiendo archivos html

```python
#EXPORTAR EL DATAFRAME FILTRADO DE PELICULAS EN UN ARCHIVO HTML
top_peliculas.to_html('top_peliculas.html', index=False)
```

```python
pd.read_html('top_peliculas.html')
```

```python
#EXPORTAR EL ARCHIVO A CSV

top_peliculas.to_csv('top_peliculas.csv', index=False)
```
